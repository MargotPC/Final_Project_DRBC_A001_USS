```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv("NAresults.csv")  # Reemplaza por el nombre correcto de tu archivo

# Agrupar por el nombre del ligando y enumerar repeticiones
df['Repetition'] = df.groupby('Ligand').cumcount() + 1

# Pivotear para que cada repetición esté en una columna
pivot = df.pivot(index='Ligand', columns='Repetition', values='Binding Affinity')

# Renombrar columnas para claridad (opcional)
pivot.columns = [f'{i}' for i in pivot.columns]

# Guardar a un nuevo CSV
pivot.to_csv("ligands_organized_NA.csv")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re

# Leer el CSV reorganizado
df = pd.read_csv("ligands_organized_NA.csv")

# Función para renombrar los ligandos
def rename_ligand(name):
    if '65028' in name:
        return 'oseltamivir'
    elif '60855' in name:
        return 'zanamivir'
    elif '154234' in name:
        return 'peramivir'
    else:
        match = re.search(r'PERUNPDB_(\d+)', name)
        if match:
            return f'peruNPDB_{match.group(1)}'
        else:
            return name


# Aplicar el renombramiento
df['Ligand'] = df['Ligand'].apply(rename_ligand)

# Establecer como índice
df.set_index("Ligand", inplace=True)

# Asegurarse que todos los valores son numéricos
df = df.apply(pd.to_numeric, errors='coerce')

# Crear el heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df, annot=True, fmt=".1f", cmap="coolwarm_r", cbar_kws={'label': 'Afinidad de Unión (kcal/mol)'})

# Etiquetas
plt.xlabel("Repeticiones")
plt.ylabel("Ligandos")
plt.xticks(rotation=0)

plt.tight_layout()
plt.savefig("heatmap_NA.png", dpi=400, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import networkx as nx

# Leer el archivo TSV
df = pd.read_csv("string_interactions.tsv", sep="\t")

# Crear grafo no dirigido
G = nx.Graph()

# Añadir las aristas a partir de las columnas de interacción
for _, row in df.iterrows():
    G.add_edge(row['#node1'], row['node2'])

# Usar la componente conectada más grande (por si hay varios subgrafos)
largest_cc_nodes = max(nx.connected_components(G), key=len)
G_largest = G.subgraph(largest_cc_nodes).copy()

# Calcular el diámetro
diameter = nx.diameter(G_largest)

# Buscar un par de nodos separados por el diámetro
source, target = None, None
for u in G_largest.nodes:
    for v in G_largest.nodes:
        if u != v and nx.shortest_path_length(G_largest, u, v) == diameter:
            source, target = u, v
            break
    if target:
        break

# Obtener la geodésica
geodesic_path = nx.shortest_path(G_largest, source=source, target=target)

# Mostrar resultados
print(f"Diámetro de la red: {diameter}")
print("Nodos que forman parte de la geodésica:")
print(geodesic_path)
```
